# Average Degree Centrality and Average Closeness Centrality

```python
centrality_df = pd.read_csv('results/centrality_metrics/degree_closeness_001.csv')
centrality_df
```

```python
sns.set_context(context='talk')

sns.relplot(x='fraction', y='graph_uncertainty', data=centrality_df)
```

```python
sns.relplot(x='fraction', y='mean_uncertainty', data=centrality_df)
```

```python
centrality_df = pd.melt(centrality_df,
        id_vars=['fraction', 'number_of_edges', 'graph_uncertainty', 'mean_uncertainty',
       'std_dev_uncertainty', 'min_uncertainty', 'max_uncertainty'],
        value_vars=['mean_degree_centrality', 'mean_closeness_centrality'],
       var_name='centrality_metric', value_name='centrality_value')
```

```python
centrality_df.head()
```

```python
sns.relplot(x='fraction', y='centrality_value', hue='centrality_metric', data=centrality_df)
```

```python
sns.relplot(x='graph_uncertainty', y='centrality_value', hue='centrality_metric', data=centrality_df)
```

```python
sns.relplot(x='mean_uncertainty', y='centrality_value', hue='centrality_metric', data=centrality_df)
```
